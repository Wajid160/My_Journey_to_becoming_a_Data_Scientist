```python
import numpy as np
import pandas as pd
```

```python
delivery = pd.read_csv("deliveries.csv")
match = pd.read_csv("matches.csv")
delivery.head() 
# Finding Top Five Runs Scorer in IPL
```

```python
runs = delivery.groupby("batsman")
```

```python
# runs.get_group("V Kohli").shape
```

```python
runs["batsman_runs"].sum().sort_values(ascending = False).head()
```

```python
# Find out the top five batsman who hit more fours
```

```python
#new_delivery = delivery[delivery["batsman_runs"] == 4]
mask = delivery["batsman_runs"] == 6
new_delivery = delivery[mask]
```

```python
new_delivery.shape[0] #total sixes in IPL
```

```python
new_delivery.groupby("batsman")["batsman_runs"].count().sort_values(ascending = False).head()
```

```python
mask = delivery["batsman"] == "V Kohli"
kohli = delivery[mask]
```

```python
kohli.head()
```

```python
kohli.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending = False).head(3)
```

```python
def run_scored(batsman_name):
    mask = delivery["batsman"] == batsman_name
    kohli = delivery[mask]
    return kohli.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending = False).head(1).index[0]
    
```

```python
run_scored("MS Dhoni")
```

- **isin() Function**: Checks if elements in a DataFrame or Series match values from a given list, returning True or False.

```python
# Find the most destructive death over batsman in the history of IPL
# bases of strike rate and minimum player played 200 balls in 16 to 20 overs
```

```python
death_over = delivery[delivery["over"] > 15]
```

```python
all_batsman = death_over.groupby("batsman")["batsman_runs"].count()
x = all_batsman > 200
```

```python
all_batsman[x].index
```

```python
batsman_list = all_batsman[x].index.tolist()
```

```python
final = delivery[delivery["batsman"].isin(batsman_list)]
```

```python
runs = final.groupby("batsman")["batsman_runs"].sum()
```

```python
balls = final.groupby("batsman")["batsman_runs"].count()
```

```python
sr = (runs/balls)*100
```

```python
sr.sort_values(ascending = False).head()
```

- **Merge Function**: Combines two DataFrames based on a common column or index, like joining tables in a database.

```python
# find orange cap holder in every season of IPL
```

```python
new = delivery.merge(match, left_on = "match_id", right_on = "id")
```

```python
new.groupby(["season","batsman"])["batsman_runs"].sum().sort_values( ascending = False).reset_index().drop_duplicates(subset = "season", keep = "first").sort_values("season")[["season","batsman"]]
# is called multy index series
```

- **Pivot Table**: Summarizes and reorganizes data from a DataFrame, such as showing average runs scored by each player in the IPL dataset, with players as rows and years or teams as columns.

```python
mask = delivery["batsman_runs"] == 6
six = delivery[mask]
six.shape
```

```python
pt = six.pivot_table(index = "over", columns = "batting_team", values = "batsman_runs", aggfunc = "count")
pt
```

```python
# using a heatmap to see pivot table
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt)
```
