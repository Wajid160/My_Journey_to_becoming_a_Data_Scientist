### Scatter Plots:
- The Scatter plot is the mainstay of statistical visualization. it depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about weather there is any meaningful relationship between them.

 - although, scatter plot can be plotted for both numarical as well as categorical data. it's mostly used to depict relationship between numerical data. 

- There are two ways to draw a scatter plot in seaborn
  1. **relplot** [Figure Level Function]
  2. **scatterplot** [Axes Level Function]

- **Axes-level functions:** These create a single plot on an individual set of axes (the area where the data is plotted). Examples include sns.scatterplot(), sns.lineplot(), etc. These are used for simpler plots.
  - **Example:** sns.scatterplot(data=df, x='age', y='salary')

- **Figure-level functions:** These manage the entire figure, which can contain multiple subplots. They are useful when you want to create more complex visualizations with several axes. Examples include sns.catplot() and sns.pairplot().

  - **Example:** sns.catplot(data=df, x='gender', y='salary', kind='box')

- as we will see, these visualizations can be quite helpful because they use simple and easily understood representations of data even on complex data by mapping upto three additional variables using the semantics of **hue, size, and style**.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

- The command plt.style.use("fivethirtyeight") applies the FiveThirtyEight style to your Matplotlib plots, giving them a specific aesthetic similar to the visual style used by the FiveThirtyEight website.

```python
plt.style.use("fivethirtyeight")
```

```python
# data = sns.load_dataset('tips')
data = pd.read_csv('tips.csv')
```

```python
data.head()
```

### Plot a Scatter Plot between total_bill and tip

- relplot() in Seaborn creates a relational plot, which shows relationships between two variables using scatter plots or line plots, while also allowing for easy grouping and customization with parameters like hue, col, or size.

```python
sns.relplot(x = 'total_bill', y = 'tip', data = data)
# by using kind equal to scatterplot it will be same 
sns.relplot(x = 'total_bill', y = 'tip', kind = 'scatter',data = data)
```

*Semantic refers to meaning or interpretation. In data visualization, semantic variables (like hue, style, or size in Seaborn) help differentiate data by encoding additional information through color, style, or size, making it easier to interpret patterns or categories in the plot.*

###  The hue Parameter:
- The hue parameter in Seaborn is used to group data points by a specific variable, applying different colors to distinguish between the categories of that variable in a plot.

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = data)
```

### The Style Parameter:
- The style parameter in Seaborn controls the appearance of data points (e.g., markers, lines) based on a categorical variable, allowing different styles (such as marker types or line styles) to represent different categories in the plot.

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'sex',data = data)
```

### The size Parameter:
- The size parameter in Seaborn adjusts the size of the data points based on a numeric or categorical variable, making it easier to represent additional information visually through varying point sizes in the plot.

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'sex',size = 'size', data = data)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'sex',size = 'size', sizes = (15, 200),data = data)
```

### Using the ScatterPlot Function:

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'sex',size = 'size', data = data)
```

## Plotting with Categorical Data:

### Catplot [Figure level Function]

### 1. Categorical Scatterplot:
- A categorical scatterplot in Seaborn visualizes relationships between a numerical variable and a categorical variable. Examples include stripplot() and swarmplot(), which display individual data points for each category.

 - **a. Stripplot**
   - a. stripplot(): Displays data points for a categorical variable with optional jitter to prevent overlap, but points can still stack on top of each other.
 - **b. Swarmplot**
   - b. swarmplot(): Displays data points while automatically adjusting them to avoid overlap, providing a clearer view without stacking.

```python
# x-axis should be categorical data
# y-axis should be numerical data 
```

```python
plt.style.use('fivethirtyeight')
```

```python
data.head()
```

```python
sns.catplot(x = 'day', y = 'tip', kind  = 'strip', data= data)
```

```python
# jitter adds slight random noise to data points along the categorical axis to prevent overlapping, making individual points easier to see.
sns.catplot(x = 'day', y = 'tip', kind  = 'strip',jitter=3 ,data= data) 
# default jitter is 1
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'swarm', data = data)
```

```python
sns.swarmplot(x='day', y = 'tip', data = data)
# we can plot like this swarmplot
```

```python
# we can only use hue
sns.catplot(x='day', y='tip', hue='sex', data=data)
```

### Box Plot:
- a boxplot is a standarized way of displaying the distribution of data based on a five number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). it can tell you about your outliers and what their values are. it can also tell your data is symmetrical. how tightly your data is grouped. and how your data is skewed.

```python
# univariant analysis
sns.boxplot(data['tip'])
```

```python
sns.boxplot(data['total_bill'])
```

```python
# multivariate analysis
sns.catplot(x='day', y='total_bill', kind = 'box',data=data)
```

```python
sns.catplot(x = 'day', y = 'total_bill', hue= 'sex', kind = 'box', data=data)
```

### Violin Plots:
- violin plots are a methid of plotting numeric data and can be considered a combination of the box plot with a kernel density plot.
- The Advantage of the violin plot over the box plot is that aside from showing the statistics it also shows the entire distribution of the data.

- violin plots are often used to compare the distribution of the given variable across some categories.

- **KDE** A kernel density plot (KDE plot) is a smooth, continuous estimate of the probability density function of a random variable, created by placing a kernel (a smooth, bell-shaped curve) at each data point and summing them up, which helps visualize the distribution of data more clearly than a histogram.

```python
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(x= 'day', y = 'total_bill', kind = 'violin', data = data)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin',hue= 'sex', data=data)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin',hue= 'sex',split= True ,data=data)
```

### Heatmaps
- A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors

```python
#flights = sns.load_dataset('flights')
flights = pd.read_csv('flights.csv')
```

```python
flights.head()
```

```python
x = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc= 'sum')
# your can remove cbar line, (cnar=False)
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar= False, linewidths=0.5, annot= True, fmt='d')
```

```python
sns.heatmap(x, cbar= False, linewidths=0.5, annot= True, fmt='d', cmap= 'summer')
#there is winter, viridis. cmaps
```

### Clustermap:

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric = 'correlation', z_score = 0, annot = True)
```

```python
sns.clustermap(x, metric = 'correlation', z_score = 0, annot = True, row_cluster=False)
```

### Jointplot:
- Jointplot creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on seperate axes

- we can use jointplots in the following 4 manners:
    1. ScatterPlots
    2. Hexbin Plots
    3. KdePlots
    4. RegPlots


```python
sns.jointplot(x='total_bill', y='tip', data=data)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=data)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=data)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=data)
```

### Displots:
- Distplots are used to univariate analysis on a continuous/numerical data.

- Distplot combines 3 things in one graph:
    1. Histogram
    2. Kernel Density Enstimation (PDF)
    3. Rugplot

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
 titanic['Age'].fillna(titanic['Age'].mean(), inplace = True)
```

### Plotting a Distplot:

```python
sns.distplot(titanic['Age'])
```

```python
sns.distplot(titanic['Age'], kde = False)
# we can use (bins = anynumber) like below
# sns.distplot(titanic['Age'], kde = False, bins = 10)
```

```python
sns.distplot(titanic['Age'], hist= False)
```

```python
sns.distplot(titanic['Age'], kde = False, hist = False, rug = True)
```
